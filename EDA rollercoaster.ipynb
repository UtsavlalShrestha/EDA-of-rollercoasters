### Data preparation ###

```python
#only keeping necessary columns and rather than dropping, copying the columns mentioned below as new df
df = df[['coaster_name',
    # 'Length', 'Speed',
    'Location', 'Status',
    # 'Opening date',
    #   'Type',
    'Manufacturer',
#     'Height restriction', 'Model', 'Height',
#        'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
#        'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
#        'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
#        'Track layout', 'Fastrack available', 'Soft opening date.1',
#        'Closing date',
#     'Opened', 
    # 'Replaced by', 'Website',
#        'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
#        'Single rider line available', 'Restraint Style',
#        'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced',
        'latitude', 'longitude',
    'Type_Main',
       'opening_date_clean',
    #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph', 
    #'height_value', 'height_unit',
    'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df.head()
```

```python
df.dtypes
```

```python
#here we can see opening_date_clean is object now we convert it into date time datatype
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
```

```python
df.dtypes
```

```python
#renaming our columns
df = df.rename(columns = { 'Location': 'location',  'Status': 'status', 'Manufacturer': 'manufacturer', 
                          'Type_Main': 'type_Main', 'Inversions_clean': 'Inversions_clean',  'Gforce_clean' :'gforce_clean'})
```

```python
#here we see only year is diff other same2same
#so what we do is take those value which doesnt have the same data in 2/3 columns
df.duplicated(subset=['coaster_name','location' ,'opening_date_clean']).sum()
```

```python
#we tae inverse of the above value so we get duplicate free value based oon those set conditions
df = df.loc[~df.duplicated(subset=['coaster_name','location' ,'opening_date_clean'])].copy()
```
